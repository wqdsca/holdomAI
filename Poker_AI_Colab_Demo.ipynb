## ğŸ§  2. í¬ì»¤ AI í•µì‹¬ ì‹œìŠ¤í…œ

```python
# í•œêµ­ì–´ í¬ì»¤ ì„¤ëª… ìƒì„±ê¸°
class KoreanPokerExplainer:
    """í•œêµ­ì–´ í¬ì»¤ ì„¤ëª… ìƒì„±ê¸°"""
    
    def __init__(self):
        self.hand_descriptions = {
            'Premium Pair': 'í”„ë¦¬ë¯¸ì—„ í˜ì–´ - ì´ëŸ° í•¸ë“œëŠ” ìì£¼ ì˜¤ì§€ ì•Šìœ¼ë‹ˆ ìµœëŒ€í•œ í™œìš©í•´ì•¼ í•´ìš”!',
            'Premium Ace': 'ê°•í•œ ì—ì´ìŠ¤ - ì–´ê·¸ë ˆì‹œë¸Œí•˜ê²Œ í”Œë ˆì´í•  ê°€ì¹˜ê°€ ìˆì–´ìš”.',
            'Medium Pair': 'ì¤‘ê°„ ê°•ë„ì˜ í˜ì–´ - ìƒí™©ì— ë”°ë¼ ì‹ ì¤‘í•˜ê²Œ í”Œë ˆì´í•˜ì„¸ìš”.',
            'Speculative Hand': 'íˆ¬ê¸°ì ì¸ í•¸ë“œ - í¬ì§€ì…˜ê³¼ ìƒí™©ì„ ì˜ ê³ ë ¤í•´ì„œ í”Œë ˆì´í•˜ì„¸ìš”.'
        }
        
        self.position_descriptions = {
            'UTG': 'UTG (ì–¸ë” ë” ê±´) - ê°€ì¥ ë¨¼ì € ì•¡ì…˜í•´ì•¼ í•˜ëŠ” ë¶ˆë¦¬í•œ ìë¦¬ì…ë‹ˆë‹¤.',
            'MP': 'MP (ë¯¸ë“¤ í¬ì§€ì…˜) - ì¤‘ê°„ ìë¦¬ë¡œ ì ë‹¹í•œ ìˆ˜ì¤€ì˜ ì–´ê·¸ë ˆì‹œë¸Œë‹ˆìŠ¤ê°€ í•„ìš”í•´ìš”.',
            'CO': 'CO (ì»·ì˜¤í”„) - ì¢‹ì€ í¬ì§€ì…˜ì´ì—ìš”! ë ˆì¸ì§€ë¥¼ ì¡°ê¸ˆ ë„“í˜€ì„œ í”Œë ˆì´í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.',
            'BTN': 'BTN (ë²„íŠ¼) - ìµœê³ ì˜ í¬ì§€ì…˜! ê°€ì¥ ë§ˆì§€ë§‰ì— ì•¡ì…˜í•˜ë¯€ë¡œ ì–´ê·¸ë ˆì‹œë¸Œí•˜ê²Œ ê°ˆ ìˆ˜ ìˆì–´ìš”.',
            'SB': 'SB (ìŠ¤ëª° ë¸”ë¼ì¸ë“œ) - ì´ë¯¸ ëˆì„ ë„£ì—ˆì§€ë§Œ í¬ì§€ì…˜ì´ ì•ˆ ì¢‹ì•„ì„œ ì¡°ì‹¬í•´ì•¼ í•´ìš”.',
            'BB': 'BB (ë¹… ë¸”ë¼ì¸ë“œ) - ë§ˆì§€ë§‰ì— ì•¡ì…˜í•  ìˆ˜ ìˆì§€ë§Œ í¬ìŠ¤íŠ¸í”Œëì—ì„œëŠ” ë¶ˆë¦¬í•©ë‹ˆë‹¤.'
        }
    
    def generate_hand_explanation(self, hole_cards, position, hand_strength, board=None, action='Bet/Raise', reasoning='ê°•í•œ í•¸ë“œë¡œ ë°¸ë¥˜ë¥¼ ì–»ì–´ì•¼ í•©ë‹ˆë‹¤'):
        """í•¸ë“œ ìƒí™©ì— ëŒ€í•œ ìì—°ì–´ ì„¤ëª… ìƒì„±"""
        
        hand_display = f"{hole_cards[0]} {hole_cards[1]}"
        
        explanation = f"""
ğŸ´ **ìƒí™© ë¶„ì„: {hand_display}**

ğŸ”¥ **{hand_display}** - {self.hand_descriptions.get(hand_strength, 'ì´ í•¸ë“œë¡œëŠ” ì¡°ì‹¬ìŠ¤ëŸ½ê²Œ ì ‘ê·¼í•˜ì„¸ìš”.')}

ğŸ‘¥ **{position}**: {self.position_descriptions.get(position, f'í¬ì§€ì…˜ {position}ì—ì„œëŠ” í¬ì§€ì…˜ì„ ê³ ë ¤í•œ í”Œë ˆì´ê°€ í•„ìš”í•´ìš”.')}
"""
        
        if board:
            board_display = ' '.join(board)
            suits = [card[1] for card in board[:3]]
            is_flush_draw = len(set(suits)) <= 2
            
            if is_flush_draw:
                explanation += f"\nğŸŒŠ **ë³´ë“œ: {board_display}** - í”ŒëŸ¬ì‹œ ë“œë¡œìš°ê°€ ìˆëŠ” ìœ„í—˜í•œ ë³´ë“œì˜ˆìš”. ë°¸ë¥˜ë² íŒ…í•  ë•ŒëŠ” í¬ê²Œ, ë¸”ëŸ¬í•‘í•  ë•ŒëŠ” ì¡°ì‹¬í•´ì•¼ í•´ìš”."
            else:
                explanation += f"\nğŸœï¸ **ë³´ë“œ: {board_display}** - ë“œë¼ì´í•œ ë³´ë“œë„¤ìš”. ìƒëŒ€ë°©ì´ ë§ì€ ê²Œ ë³„ë¡œ ì—†ì„ ê°€ëŠ¥ì„±ì´ ë†’ì•„ì„œ ë¸”ëŸ¬í•‘í•˜ê¸° ì¢‹ì€ ìƒí™©ì´ì—ìš”."
        
        explanation += f"""

---

ğŸš€ **ì¶”ì²œ ì•¡ì…˜: {action}**

**ì´ìœ **: {reasoning}

ì´ ìƒí™©ì—ì„œëŠ” {action}ì´ ê°€ì¥ ì´ë¡ ì ìœ¼ë¡œ ì˜¬ë°”ë¥¸ ì„ íƒì…ë‹ˆë‹¤.
ë¬¼ë¡  ìƒëŒ€ë°©ì˜ ì„±í–¥ì´ë‚˜ ê²Œì„ í”Œë¡œìš°ì— ë”°ë¼ ì¡°ì •í•  ìˆ˜ ìˆì–´ìš”.

---

ğŸ¨ **ìŠ¤íƒ€ì¼ë³„ ì ‘ê·¼ë²•**

â€¢ **TAG (íƒ€ì´íŠ¸-ì–´ê·¸ë ˆì‹œë¸Œ)**: ë³´ìˆ˜ì ì´ì§€ë§Œ ë°¸ë¥˜ ì¤‘ì‹¬ì˜ í”Œë ˆì´
â€¢ **LAG (ë£¨ì¦ˆ-ì–´ê·¸ë ˆì‹œë¸Œ)**: ì–´ê·¸ë ˆì‹œë¸Œí•˜ê³  ë¸”ëŸ¬í”„ë¥¼ ë§ì´ í™œìš©
â€¢ **Balanced**: ìˆ˜í•™ì ìœ¼ë¡œ ìµœì í™”ëœ GTO ê¸°ë°˜ í”Œë ˆì´

ê°ìì˜ ì„±í–¥ê³¼ ì‹¤ë ¥ì— ë§ëŠ” ìŠ¤íƒ€ì¼ì„ ì„ íƒí•˜ì„¸ìš”! ğŸ¯
"""
        
        return explanation.strip()

# í¬ì»¤ ë¶„ì„ ì—”ì§„
class PokerAnalyzer:
    """í¬ì»¤ ìƒí™© ë¶„ì„ ì—”ì§„"""
    
    def evaluate_hand_strength(self, hole_cards, board=None):
        """í•¸ë“œ ê°•ë„ í‰ê°€"""
        if not board:  # í”„ë¦¬í”Œë
            if hole_cards[0][0] == hole_cards[1][0]:  # í˜ì–´
                return "Premium Pair" if hole_cards[0][0] in 'AKQJ' else "Medium Pair"
            elif hole_cards[0][0] in 'AK' and hole_cards[1][0] in 'AK':
                return "Premium Ace"
            else:
                return "Speculative Hand"
        else:
            return "Made Hand"
    
    def get_recommended_action(self, hand_strength, position):
        """ì¶”ì²œ ì•¡ì…˜ ê³„ì‚°"""
        if hand_strength in ['Premium Pair', 'Premium Ace']:
            return 'Bet/Raise', 'ê°•í•œ í•¸ë“œë¡œ ë°¸ë¥˜ë¥¼ ì–»ì–´ì•¼ í•©ë‹ˆë‹¤'
        elif hand_strength == 'Medium Pair':
            if position in ['BTN', 'CO']:
                return 'Bet/Raise', 'ì¢‹ì€ í¬ì§€ì…˜ì—ì„œ ë°¸ë¥˜ë¥¼ ë…¸ë ¤ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤'
            else:
                return 'Check/Call', 'ì¤‘ê°„ ê°•ë„ í•¸ë“œë¡œ íŒŸ ì»¨íŠ¸ë¡¤ì´ í•„ìš”í•©ë‹ˆë‹¤'
        else:
            return 'Check/Fold', 'ì•½í•œ í•¸ë“œë¡œ ë³´ìˆ˜ì ì¸ í”Œë ˆì´ê°€ í•„ìš”í•©ë‹ˆë‹¤'

print('ğŸ§  í¬ì»¤ AI í•µì‹¬ ì‹œìŠ¤í…œ ë¡œë“œ ì™„ë£Œ!')
```

## ğŸ¯ 3. ì‹¤ì œ í¬ì»¤ ìƒí™© ë¶„ì„ ë°ëª¨

```python
# í¬ì»¤ AI ì‹œìŠ¤í…œ ì´ˆê¸°í™”
explainer = KoreanPokerExplainer()
analyzer = PokerAnalyzer()

print('ğŸ¯ í¬ì»¤ AI í”Œë ˆì´ ìŠ¤íƒ€ì¼ ê°€ì´ë“œ ì‹œìŠ¤í…œ')
print('=' * 60)

# ì˜ˆì‹œ ìƒí™©ë“¤
situations = [
    {
        'hole_cards': ['As', 'Kd'],
        'position': 'BTN',
        'board': ['Qh', '7c', '2s'],
        'description': 'ãƒ—ãƒªãƒŸã‚¢ãƒ  ã‚¨ãƒ¼ã‚¹ vs ãƒ‰ãƒ©ã‚¤ãƒœãƒ¼ãƒ‰'
    },
    {
        'hole_cards': ['Jh', 'Js'],
        'position': 'UTG',
        'board': None,
        'description': 'ãƒ—ãƒªãƒ•ãƒ­ãƒƒãƒ— ãƒ—ãƒªãƒŸã‚¢ãƒ ãƒšã‚¢'
    },
    {
        'hole_cards': ['8c', '7c'],
        'position': 'CO',
        'board': ['9h', '6s', '2d'],
        'description': 'ã‚¹ãƒšã‚­ãƒ¥ãƒ¬ã‚¤ãƒ†ã‚£ãƒ–ãƒãƒ³ãƒ‰ vs ãƒ‰ãƒ­ãƒ¼'
    }
]

for i, situation in enumerate(situations, 1):
    print(f'\nğŸ´ **ìƒí™© {i}**: {situation["description"]}\n')
    
    # í•¸ë“œ ê°•ë„ ë¶„ì„
    hand_strength = analyzer.evaluate_hand_strength(
        situation['hole_cards'], 
        situation['board']
    )
    
    # ì¶”ì²œ ì•¡ì…˜ ê³„ì‚°
    action, reasoning = analyzer.get_recommended_action(
        hand_strength, 
        situation['position']
    )
    
    # í•œêµ­ì–´ ì„¤ëª… ìƒì„±
    explanation = explainer.generate_hand_explanation(
        hole_cards=situation['hole_cards'],
        position=situation['position'],
        hand_strength=hand_strength,
        board=situation['board'],
        action=action,
        reasoning=reasoning
    )
    
    print(explanation)
    print('\n' + '='*60)

print('\nâœ… í¬ì»¤ ìƒí™© ë¶„ì„ ì™„ë£Œ! í•œêµ­ì–´ë¡œ ì™„ë²½í•˜ê²Œ ì„¤ëª…ë©ë‹ˆë‹¤! ğŸš€')
```

## ğŸ—ºï¸ 6. ê°œì¸ ë§ì¶¤ í•™ìŠµ ë¡œë“œë§µ

```python
# í•™ìŠµ ë¡œë“œë§µ ìƒì„±ê¸°
def generate_learning_roadmap(current_stats, target_style='TAG'):
    """ê°œì¸ ë§ì¶¤ í•™ìŠµ ë¡œë“œë§µ ìƒì„±"""
    
    target_profiles = {
        'TAG': {'vpip': 0.22, 'pfr': 0.18, 'af': 3.5},
        'LAG': {'vpip': 0.32, 'pfr': 0.28, 'af': 4.2},
        'Balanced': {'vpip': 0.25, 'pfr': 0.20, 'af': 2.8}
    }
    
    target = target_profiles[target_style]
    
    print(f'ğŸ¯ **{target_style} ìŠ¤íƒ€ì¼ í•™ìŠµ ë¡œë“œë§µ**')
    print('=' * 60)
    
    print(f"\nğŸ“Š **í˜„ì¬ í”Œë ˆì´ ìŠ¤íƒ€ì¼ ë¶„ì„**")
    print(f"â€¢ VPIP (í”Œë ˆì´í•˜ëŠ” í•¸ë“œ ë¹„ìœ¨): {current_stats['vpip']*100:.1f}%")
    print(f"â€¢ PFR (í”„ë¦¬í”Œë ë ˆì´ì¦ˆ ë¹„ìœ¨): {current_stats['pfr']*100:.1f}%")
    print(f"â€¢ ì–´ê·¸ë ˆì‹œë¸Œ íŒ©í„°: {current_stats['af']:.1f}")
    
    # ê°œì„  ì˜ì—­ ì‹ë³„
    improvements = []
    
    vpip_diff = abs(target['vpip'] - current_stats['vpip'])
    if vpip_diff > 0.05:
        direction = "ë” ë§ì€" if target['vpip'] > current_stats['vpip'] else "ë” ì ì€"
        improvements.append({
            'area': 'Hand Selection (VPIP)',
            'current': f"{current_stats['vpip']:.1%}",
            'target': f"{target['vpip']:.1%}",
            'action': f"{direction} í•¸ë“œ í”Œë ˆì´í•˜ê¸°",
            'priority': 'High' if vpip_diff > 0.1 else 'Medium'
        })
    
    pfr_diff = abs(target['pfr'] - current_stats['pfr'])
    if pfr_diff > 0.03:
        direction = "ë” ìì£¼" if target['pfr'] > current_stats['pfr'] else "ëœ ìì£¼"
        improvements.append({
            'area': 'Preflop Aggression (PFR)',
            'current': f"{current_stats['pfr']:.1%}",
            'target': f"{target['pfr']:.1%}",
            'action': f"{direction} ë ˆì´ì¦ˆí•˜ê¸°",
            'priority': 'High'
        })
    
    af_diff = abs(target['af'] - current_stats['af'])
    if af_diff > 0.5:
        direction = "ë” ì–´ê·¸ë ˆì‹œë¸Œí•˜ê²Œ" if target['af'] > current_stats['af'] else "ë” ë³´ìˆ˜ì ìœ¼ë¡œ"
        improvements.append({
            'area': 'Postflop Aggression (AF)',
            'current': f"{current_stats['af']:.1f}",
            'target': f"{target['af']:.1f}",
            'action': f"{direction} ë² íŒ…/ë ˆì´ì¦ˆí•˜ê¸°",
            'priority': 'Medium'
        })
    
    # ê°œì„  ì‚¬í•­ ì¶œë ¥
    print(f"\nğŸš€ **ê°œì„ ì´ í•„ìš”í•œ ë¶€ë¶„**")
    for i, improvement in enumerate(improvements, 1):
        priority_emoji = "ğŸ”¥" if improvement['priority'] == 'High' else "â­"
        print(f"{priority_emoji} **{improvement['area']}**")
        print(f"   í˜„ì¬: {improvement['current']} â†’ ëª©í‘œ: {improvement['target']}")
        print(f"   ì•¡ì…˜: {improvement['action']}\n")
    
    # ì‹¤ì²œ ë°©ë²•
    print(f"ğŸ’¡ **êµ¬ì²´ì ì¸ ì‹¤ì²œ ë°©ë²•**")
    print(f"""    
**1ì£¼ì°¨: í•¸ë“œ ì„ íƒ ê°œì„ **
- í¬ì§€ì…˜ë³„ ë ˆì¸ì§€ ì°¨íŠ¸ë¥¼ ë§Œë“¤ì–´ì„œ ì°¸ê³ í•˜ì„¸ìš”
- ê²Œì„ ì¤‘ì— "ì´ í•¸ë“œë¥¼ ì •ë§ í”Œë ˆì´í•´ì•¼ í•˜ë‚˜?" ìë¬¸í•˜ê¸°
- í•˜ë£¨ì— 100í•¸ë“œì”© í”Œë ˆì´í•˜ë©° í•¸ë“œ ì„ íƒ ì—°ìŠµ

**2-3ì£¼ì°¨: ì–´ê·¸ë ˆì‹œë¸Œë‹ˆìŠ¤ ì¡°ì •**
- ì¢‹ì€ í•¸ë“œë¥¼ ê°€ì¡Œì„ ë•ŒëŠ” ë²¨ë¥˜ë¥¼ ì–»ê¸° ìœ„í•´ ë² íŒ…í•˜ê¸°
- ë¸”ëŸ¬í•‘í•  ë•ŒëŠ” ìŠ¤í† ë¦¬ê°€ ìˆëŠ”ì§€ ìƒê°í•´ë³´ê¸°
- C-bet ë¹ˆë„ë¥¼ ì¡°ê¸ˆì”© ëŠ˜ë ¤ê°€ê¸°

**4ì£¼ì°¨ ì´í›„: ê³ ê¸‰ ê°œë… ì ìš©**
- ìƒëŒ€ë°© íƒ€ì…ì— ë”°ë¥¸ ì „ëµ ì¡°ì • ì—°ìŠµ
- ë² íŒ… ì‚¬ì´ì§• ìµœì í™”
- í•¸ë“œ ë¦¬ë·°ë¥¼ í†µí•œ ì§€ì†ì  ê°œì„ 
""")
    
    return improvements

# ì˜ˆì‹œ ì‚¬ìš©ì í†µê³„ë“¤
user_examples = [
    {
        'name': 'ë£¨ì¦ˆí•œ ì´ˆë³´ì',
        'stats': {'vpip': 0.45, 'pfr': 0.15, 'af': 1.8},
        'target': 'TAG'
    },
    {
        'name': 'íƒ€ì´íŠ¸í•œ ì¤‘ê¸‰ì', 
        'stats': {'vpip': 0.18, 'pfr': 0.25, 'af': 2.2},
        'target': 'Balanced'
    },
    {
        'name': 'ì–´ê·¸ë ˆì‹œë¸Œ ê³ ê¸‰ì',
        'stats': {'vpip': 0.28, 'pfr': 0.22, 'af': 3.8},
        'target': 'LAG'
    }
]

# ê° ì‚¬ìš©ìë³„ ë¡œë“œë§µ ìƒì„±
for user in user_examples:
    print(f"\nğŸ‘¤ **{user['name']} ì‚¬ë¡€**")
    improvements = generate_learning_roadmap(user['stats'], user['target'])
    print('\n' + '='*60)

print('\nâœ… ê°œì¸ ë§ì¶¤ í•™ìŠµ ë¡œë“œë§µ ìƒì„± ì™„ë£Œ!')
print('ğŸ¯ ê°ìì˜ í˜„ì¬ ìˆ˜ì¤€ì— ë§ëŠ” êµ¬ì²´ì ì¸ ê°œì„  ë°©í–¥ì„ ì œì‹œí•©ë‹ˆë‹¤!')
```
